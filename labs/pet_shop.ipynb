For this diagram I created six tables. 

Since this is a shop, I created a table for customers. This includes the customer's phone number, address, first and last name, and loyalty ID as a unique identifier (used as primary key). These are all important attributes in case there is a need to contact the customer.

Since this is a shop, I created two individual tables for its product transactions: pet and food transactions. The pet transactions table includes the price of the pet, the pet's unique ID (from Pets table, used as foreign key), and the customer's loyalty ID (from Customer table, used as foreign key). The foods transactions table includes the customer's loyalty ID, the food product ID, and price of the food. These are important attributes for a basic transaction in a pet shop.

I created the adoption table in conjunction with the pet transactions table because if a customer goes through a pet transaction, then he/she has therefore adopted a pet. The Adoption table includes the most of what the Customer table includes, first and last name, phone number, home address, as well as the pet ID as this is vital information to the shop so it can keep track of who has bought what pet and how to contact the customer.

Lastly, I created the Pets and Food table as these are the main products that make up the shop. The Pets table includes the pet species, breed, and in what enclosure it can be located (via ID). The Food table includes the food type, price, and brand as well as the food product ID. This is important for transactions and because it is used to keep inventory and analyze sales.
